```python
import os
import numpy as np
import matplotlib.pyplot as plt
from screaming_rocks.genData import dataLoader
%matplotlib inline


d = '../seismology_data/'
files = os.listdir(d)
files.sort()
pairs = []
for i, f in enumerate(files[:-1]):
    if f[:-4] == files[i+1][:-4]:
        pairs.append(f[:-4])
print(pairs)
```

```python
pair = dataLoader(d+'rct-uop-021115.data.00015.wve',
                  d+'rct-uop-021115.data.00015.srm',
                  chunk_time=0.001)
print(pair.duration)
data = pair[758]
```

```python
fig, ax = plt.subplots(nrows=pair.nchannels, figsize=(16, 16), sharex=True)
fig.subplots_adjust(hspace=0.0)
for i in range(pair.nchannels):
    ax[i].plot(data[:, i])
```

```python
freq, psds = pair.welch()
fig, ax = plt.subplots(nrows=pair.nchannels, figsize=(16, 16), sharex=True)
fig.subplots_adjust(hspace=0.0)
for i in range(pair.nchannels):
    ax[i].loglog(freq, psds[:, i])
```

```python
white = pair.whitened(758)
fig, ax = plt.subplots(nrows=pair.nchannels, figsize=(16, 16), sharex=True)
fig.subplots_adjust(hspace=0.0)
for i in range(pair.nchannels):
    ax[i].plot(white[:, i])
```
